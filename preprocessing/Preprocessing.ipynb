```python
import pandas as pd
import re
```

```python
crawlingdata = pd.read_csv("/media/reindrairawan/New Volume/Data E/REPOSITORY/PYTHON/ProjectKuliah/NLP/ProjectUas(Sentimen_Analisis)/crawlingdata/Scbd fashion week.csv", names = ["text"])
crawlingdata
```

```python
crawlingdata.drop_duplicates(subset='text',inplace=True)
crawlingdata.describe()
```

```python
def casefolding(text):
    text = text.lower() # merubah bentuk menjadi lower case
    text = re.sub(r'\d+', '', text) # menghapus angka
    text = re.sub(r'[^\w\s]','',text) # menghapus karakter tanda baca
    # text = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",text).split())

    return text


crawlingdata['casefolding'] = crawlingdata['text'].apply(lambda x: casefolding(x))
crawlingdata
```

```python

key_norm = pd.read_csv("/media/reindrairawan/New Volume/Data E/REPOSITORY/PYTHON/ProjectKuliah/NLP/ProjectUas(Sentimen_Analisis)/filePreprocessing/clean_data.csv")

def text_normalize(text):
    text = ' '.join([key_norm[key_norm["singkat"] == word]['hasil'].values[0] if (key_norm["singkat"] == word).any() else word for word in text.split()])
    text = str.lower(text)
    return text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stop_factory = stopwords.words('indonesian')
more_stopword = ['daring', 'online', 'pd']

def remove_stop_words(text):
    sw = stop_factory + more_stopword
    clean_words = []
    text = text.split()
    for word in text:
        if word not in sw:
            clean_words.append(word)
    return " ".join(clean_words)


crawlingdata['remove stop words'] = crawlingdata['text'].apply(lambda x: remove_stop_words(x))
crawlingdata.head()
```

```python
# stemming menggunakan sastrawi 
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(text):
    text = stemmer.stem(text)
    return text

 
crawlingdata['stemming'] = crawlingdata['text'].apply(lambda x: stemming(x))
crawlingdata.head()  
```

```python
# Text Preprocessing Pipeline
#opsional 
def data_process(text):
    text = casefolding(text)
    # text = text_normalize(text)
    text = remove_stop_words(text)
    text = stemming(text)
    return text
```

```python
import time

t1 = time.time()
crawlingdata['output'] = crawlingdata['text'].apply(data_process)
print('cleaning dataset elapsed time:', time.time() - t1, '\n')
```
